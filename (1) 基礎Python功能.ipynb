### 13 jupyter notebook互動功能應用於畫圖

```python
def myplot(n=1):
    x = np.linspace(-5, 5, 500)
    #sinc函數
    y = np.sinc(n*x)
    plt.plot(x, y)
```

```python
myplot()
```

```python
interact(myplot, n=(1, 20));
```

### 15 while迴圈
* 無限迴圈 : 若符合條件會一直執行，所以要設停損點

```python
s = 0
while s<10:
    print(s)
    s = s + 1
```

```python
msg = "hello"

while (msg != "bye"):
    msg = input(">> ")
    print("拍拍!")
```

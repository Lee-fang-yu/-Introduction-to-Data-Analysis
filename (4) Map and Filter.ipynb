## 1 Map : 每個在list中的元素都需要做計算

### e.g. 美元換台幣
### (1) 基本做法

```python
items = [500, 30, 2000]
c = 30.8594
```

```python
egg = [] 
for i in items:
    p = i*c
    egg.append(p)
```

```python
egg
```

### (2) map函數

```python
def exchange(p):
    return p*c
```

```python
#exchange(500)
```

```python
list(map(exchange, items))
```

### (3) lambda函數

```python
list(map(lambda p:p*c, items))
```

### (4) list comprehension

```python
[i*c for i in items]
```

## 2 Filter : 過濾出某條件之下的東西

### e.g. 過濾及格成績
### (1) 基本作法

```python
grades = [50, 59, 65, 80, 100, 35, 40, 88]
```

```python
def spam(g):
    return g >= 60
```

```python
#spam(50)
#spam(99)
```

```python
#過濾出來
list(filter(spam, grades))
```

### (2) lambda函數

```python
list(filter(lambda g:g>=60, grades))
```

## 3 print的使用

```python
for i in range(10):
    print(i)
```

### ● 用空格隔開

```python
for i in range(10):
    print(i, end = ' ')
```

### ● 用逗點隔開

### method 1

```python
#最後一個逗點多出來啦
for i in range(10):
    print(i, end = ',')
```

### method 2 

```python
#str應該是string的意思
hamm = list(map(str, range(10)))
```

```python
print(','.join(hamm))
```

## 4 print格式化

### 4.1 希望能一次將三種形式的東西加起來

### e.g.

```python
name = '87'
age = 25
```

```python
print('您好, 我是'+name+'今年'+str(age)+'歲。')
```

#### method 1
* name : %s 字串 
* age : %d 數字

```python
egg = "您好, 我是%s, 今年%d歲。"%(name, age)
```

```python
print(egg)
```

####  method 2 : format

```python
egg = "您好, 我是{}, 今年{}歲。".format(name, age)
```

```python
print(egg)
```

#### method 3 : f-string

```python
egg = f"您好, 我是{name}, 今年{age}歲。"
```

```python
print(egg)
```

### 4.2 數字意思顯示

```python
c = 30.8594353
```

```python
print(f"1美金合台幣{c}元。")
```

#### ● {c:.2f}中的.2f表取到小數點後2位且為浮點數的形式

```python
print(f"1美金合台幣{c:.2f}元。")
```

#### ●{c:10.2f}中的10.2f表空10格且取到小數點後2位且為浮點數的形式

```python
print(f"1美金合台幣{c:10.2f}元。")
```

### 4.3 圖示的顯示

```python
egg = ":)"
spam = f"●{egg}●"
print(spam)
```

#### round 1
* 20s:空20格要加s因為是string

```python
spam = f"●{egg:20s}●"
print(spam)
```

#### round 2

```python
spam = f"●{egg:>20s}●"
print(spam)
```

#### round 3
* ^ :置中

```python
spam = f"●{egg:^20s}●"
print(spam)
```

#### round 4

```python
spam = f"●{egg:*^20s}●"
print(spam)
```

### e.g. 畫聖誕樹

```python
level = 10
for n in range(level):
    s = "*"*(2*n+1)
    print(f"{s:^21s}")
```

### 4.4數字的顯示---補0

#### round 5

```python
n = 87

spam = f"{str(n):0>3s}"#所以87靠右邊的
print(spam)
```

```python
n = 87

spam = ("000"+str(n))[-3:]

print(spam)
```

```python
n = str(87)

#不足的地方補0
spam = n.zfill(3)
print(spam)
```
